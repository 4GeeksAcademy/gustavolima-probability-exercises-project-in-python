### Exercise 1 

Two unbiased dice are thrown once and the total score is observed. Use a simulation to find the estimated probability that the total score is even or greater than 7

We could use a mathematical equation to solve this or we can run the experiment many times, and then count the number of times where the score is even or greater than 7

-Run the experiment 1000 times (roll 2 dice 1000 times, and sum the result)

-Keep track of the number of times that the sum was either greater than 7 or an even number

-Divide the number from step 2 by the number of iterations (1000)


```python
#import libraries
import numpy as np

#Create a Counting Variable
count = 0

#Define number of rolls
rolls = 1000

# Create a Loop to roll 1000 times the dices
for i in range(rolls):
    dice1 = np.random.randint(1,6)
    dice2 = np.random.randint(1,6)

    # Score Calculations
    total = dice1 + dice2

    # Conditions to be met for counting
    if total %2 == 0 or total > 7:
        count += 1

# Print the % of such events to happen
print(f'The Probability of Rolling Even or Great than 7 is:', round((count/rolls)*100, 2) ,f'%')
```

### Exercise 2

A box contains 10 white balls, 20 reds and 30 greens. Draw 5 balls with replacement. What is the probability that:

a. 3 white or 2 red

b. All 5 are the same color


We will use the same system here. We will pick our balls during each round and count the number of times that:

a. 3 white or 2 red 

b. All 5 are the same color.

```python
import numpy as np

d = {}

for i in range(60):
   if i < 10:
      d[i] = 'White'
   elif i > 9 and i < 30:
      d[i] = 'Red'
   else:
      d[i] = 'Green'  
  
        
#Initialize important variables

countwhred = 0
countfivesame = 0
n_simulations = 1000 


for i in range(n_simulations):
    
   #Make a list of the colors that we choose
   colors = []
   
   # Run another nested Loop to pick 5 balls out of our d dictionary above
   # and add those picks the color list
   for i in range(5):
      colors.append(d[np.random.randint(0,59)])

   #Convert it to a numpy array
   colors = np.array(colors)

   #find the number of each that we picked
   white = sum(colors == 'White')
   red = sum(colors == 'Red')
   green = sum(colors == 'Green')

    
   #Keep track if the combination met the criteria
   if white == 3 and red == 2:
      countwhred += 1
   if white == 5 or red == 5 or green == 5:
      countfivesame += 1


#print 'the probability of 3 white and 2 red is' in percentage
print(f'The probability of 3 whte and 2 red is;', round((countwhred/n_simulations)*100, 2), f'%')

#print 'the probability of all the same color is' in percentage  
print(f'The probability of all same color;', round((countfivesame/n_simulations)*100, 2), f'%') 
```

```python
import numpy as np

#Initialize important variables

countwhred = 0
countfivesame = 0
runs = 1000 

# Run simulation with a for loop

for i in range(runs):
    # Using random.choice and defining p= for distributing of the balls colors (10, 20, 30)
    colors = np.random.choice(['white', 'red', 'green'], size=5, p=[0.17, 0.33, 0.50])

    # Sum the number of each ball
    white = np.sum(colors == 'white')
    red = np.sum(colors == 'red')
    
    # Define the condition to be met, 3 white + 2 red               
    if white == 3 and red == 2:
        countwhred += 1

    # Define the condition for all the same, counting at index 0
    if np.all(colors == colors[0]):
        countfivesame += 1
                  

#print 'the probability of 3 white and 2 red is' in percentage
print(f'The probability of 3 whte and 2 red is;', round((countwhred/n_simulations)*100, 2), f'%')

#print 'the probability of all the same color is' in percentage  
print(f'The probability of all same color;', round((countfivesame/n_simulations)*100, 2), f'%') 

# This solution for me seems easier to understand as we can define in one line all the balls variables and randomly
# ask it to pick 5, with replacement because is the default behaviour with np.random.choice. The rest is similiar
# as i couldn't find another way to have less code
```
